# Exercise 1
Fill in the blanks(indicated with a ?), so the code below will produce the plot shown.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='maroon', linestyle='dashed',
             marker= 'o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

# Exercise 2
Use NumPy to generate $x$ and $y$ data suitable for plotting the graph of the function:
$$
y = e^{-x/10} sin(x)
$$
for values in the range 0 to 50.


```python
x = np.linspace(0,50,1000)
y = np.e**(-(x)/10)*(np.sin(x))

plt.plot(x,y,label='$y=e^{-x/10}sin(x)$',color='tab:blue')
plt.xlabel('x',color='darkorange',fontsize='15')
plt.ylabel('y',rotation=0,color='darkorange',fontsize='15')
plt.title('$y=e^{-x/10}sin(x)$',fontsize='20')
plt.grid(alpha=.25)
plt.ylim([-1,1])
plt.show()
```
