### Things to note

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = [0,1,2,3,4,5,6]
y_1 = [0,2,4,6,8,10,12]
y_2 = [0,3,6,9,12,15,18]
y_3 = [0,4,8,12,16,20,24]
x_error, y_error = .1, 0.75         # what does this do?

plt.plot(x,y_1,label ='Y1 values', c='red', ls='dashed')
plt.plot(x,y_2,label ='Y2 values', c='blue')
plt.errorbar(x,y_3, yerr=y_error, label = 'Y3 with errors', c='green')
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 1.2 Adding another plot

```python
time = np.arange(10)
y2 = time**2
y3 = time**3

plt.plot(time,y2, label='$y^2$', linestyle='--', color='#1b9e77')
plt.plot(time,y3, label='$y^3$')

plt.xlabel('time (s)')
plt.ylabel('Hahaha (ha)')
plt.title('My First Plot')
plt.legend()
plt.grid(alpha=.25)
```

# 2 Better with NumPy

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)              # better to convert to numpy arrays
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

```python
x = np.array([0,1,2,3,4,5,6])

x2 = x**2
sin_x = np.sin(x)

plt.plot(x,x2,label='x^2',c='red',ls='dashed')
plt.plot(x,sin_x,label='sin(x)',c='blue')
plt.legend(loc='upper left')
```

## We need another axis!

```python
x = np.array([0,1,2,3,4,5,6])
x = np.linspace(0,6,100)


x2 = x**2
sin_x = np.sin(x)

plt.plot(x,x2,label='x^2',c='red',ls='dashed')
plt.legend(loc='lower left')

plt.twinx()

plt.plot(x,sin_x,label='sin(x)',c='blue')
plt.legend(loc='lower right')
```
