# 2 Comparing the two ‘dialects’

```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

### pyplot Interface

```python
plt.plot(x, sin_x, label='$sinx$')
plt.plot(x, cos_x, label='$cosx$')
plt.grid()
plt.legend()
```

### OO Interface

```python
fig, ax = plt.subplots(nrows=2,figsize=(8,4))

my_sin = ax[0]  
my_cos = ax[1]

my_sin.plot(x, sin_x, label='$\\cos x$', color='black')
my_sin.legend()

my_cos.plot(x, cos_x, label='$\\sin x$',color='black')
my_cos.legend()

my_sin.fill_between(x,0,sin_x,alpha=.5,color='pink')
my_cos.fill_between(x,0,cos_x,alpha=.5,color='pink')

plt.suptitle('Trig Functions')

my_sin.set_title('Sine Graph')
my_cos.set_title('Cosine Graph')

my_sin.set_xlabel('Angle')
my_cos.set_xlabel('Angle')

my_sin.set_ylim(-1,1)
my_cos.set_ylim(-1,1)

my_sin.set_xlim(-np.pi,np.pi)
my_cos.set_xlim(-np.pi,np.pi)

my_sin.set_ylabel('$sin(x)$',rotation=0)
my_cos.set_ylabel('$cos(x)$',rotation=0)

my_sin.grid()
my_cos.grid()

plt.show()
```

# 5 Multiple rows and columns

### Using ax

```python

fig, ax = plt.subplots(
    nrows=2,ncols=3,
    figsize=(8, 4),
    sharex='col',sharey='row'
)

for row in range(2):
    for col in range(3):
        ax[row,col].plot(x,col*cos_x,label=f'{row},{col}',color='black',alpha=.5)
        ax[row,col].fill_between(x,col*cos_x,alpha=.25,color='pink')

for a in ax.flatten():
    a.grid(alpha=.5,c='pink')
    a.legend(loc='upper left')

plt.tight_layout()
```
